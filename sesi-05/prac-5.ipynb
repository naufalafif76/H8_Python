```python
city_countries = pd.DataFrame({
    "country": ['Holland', 'Japan', 'Canada', 'Spain'],
    "capital": [1, 1, 0, 0]},
    index=['Amsterdam', 'Tokyo', 'Torento', 'Barcelona']
)
city_countries
```

```python
pd.merge(cities, countris, left_on='country', right_index=True)
```
